## 2. Data Visualization

* Ensure the visualizations are clear and informative.

### Create visualizations to illustrate the findings from the EDA.


```python
plt.figure(figsize=(10, 5))
sns.histplot(df['City'].value_counts(), bins=5, kde=True)
plt.title('Number of accedint')
plt.show()
```

```python
df.info()
```

```python
correlation_matrix = df[['TrafficIndexLive', 'JamsCount', 'JamsDelay', 'JamsLength', 'TrafficIndexWeekAgo', 'TravelTimeHistoric','TravelTimeLive']].corr()

# Heatmap for correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

### Use appropriate plots such as histograms, bar charts, pie charts, scatter plots, and heatmaps.

```python
Count_jam = df['City'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(Count_jam, labels=Count_jam.index)
plt.title('Distribution of Injuries Type')
plt.show()
```
